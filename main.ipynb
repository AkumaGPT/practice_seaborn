GETTING STARTED WITH SEABORN

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# first practice
df = pd.read_csv("true_pred.csv")
# df.isna().sum()
df_n = df.head()

sns.set_theme(style="whitegrid")

sns.scatterplot(data=df_n, x="actual", y="predicted")
plt.show()
```

```python
df = pd.read_csv("true_pred.csv")
df = df.head()

# get thier error
df.loc[:, "error"] =  (df["actual"].sub(df["predicted"])).abs()

# get the group column
df.loc[:, "quality"] = df["error"].apply(lambda e: "good" if e <= 2 else "bad")

# color dots by good or bad
sns.set_theme(style="whitegrid", font="monospace", font_scale=1.2, palette="bright")
sns.scatterplot(data=df, x="actual", y="predicted", hue="quality")

plt.show()
```

```python
# ML challenge

df = pd.read_csv("true_pred.csv")
df = df.head(6)

# get the error
df["error"] = (df["actual"] - df["predicted"]).abs()

# get the quality
df["quality"] = df["error"].apply(lambda e: "good" if e <= 2 else "bad")

# set theme and plot with hue
sns.set_theme(style="darkgrid", font="serif", font_scale=0.9)

sns.scatterplot(data=df, x="actual", y="predicted", hue="quality")

plt.show()
```

```python
# recall test
df = pd.read_csv("true_pred.csv")

# using 10 rows
df = df.head(10)

# get the error and quality
df.loc[:, "error"] = (df["actual"].sub(df['predicted'])).abs()
df["quality"] = df["error"].apply(lambda e: "good" if e <= 2 else "bad")

# plot a seaborn scatter with hue
sns.set_theme(style="whitegrid")
sns.scatterplot(data=df, x="actual", y="predicted", hue="quality")

plt.show()
```

SEABORN HEATMAP

```python
cm = [[8, 2], [1, 9]]

sns.heatmap(cm, annot=True, fmt="d", cmap="viridis")

plt.show()
```

did fmt="d", what was that?

```python
# to label axes so it looks like a real cm

cm = [[50, 5, 0],
      [3, 40, 7],
      [0, 4, 45]]

fig, ax = plt.subplots()

sns.heatmap(data=cm, annot=True, fmt="d", cmap="viridis", ax=ax)

ax.set_title("Confusion Matrix")
ax.set_xlabel("predicted")
ax.set_ylabel("actual")

plt.show()
```

```python
# recall ML challenge

cm = [[30, 2],
      [20, 8]]

fig, ax = plt.subplots()

sns.heatmap(cm, annot=True, fmt="d", linewidths=1, cmap="viridis")

ax.set_title("Confusion Matrix")
ax.set_xlabel("Actual")
ax.set_ylabel("Predicted")

plt.show()
```

Valid format code -> integers -> fmt = "d", 2 decimals -> fmt = ".2f"

```python
# recall test challenge
cm = [[80, 5, 2],
      [7, 75, 3],
      [1, 4, 70]]

sns.heatmap(cm, annot=True, fmt="d", linewidths=1, cmap="plasma")

plt.show()
```

PAIRPLOT + CORRELATION HEATMAP (FEATURE RELATIONSHIPS)

```python
df = pd.DataFrame({
    "x1": [1, 2, 3, 4, 5],
    "x2": [2, 4, 6, 8, 10],
    "y": [1, 1, 2, 3, 5]
})

sns.pairplot(df)

plt.show()
```

```python
df = pd.DataFrame({
  "x1": [1,2,3,4,5,6],
  "x2": [1,2,2,4,4,6],
  "x3": [2,1,3,2,5,4],
  "label": ["A","A","A","B","B","B"]
})

sns.pairplot(df, hue="label")

plt.show()
```

```python
df = pd.DataFrame({
  "x1": [1,2,3,4,5],
  "x2": [2,4,6,8,10],
  "x3": [5,4,3,2,1]
})

corr = df.corr()

sns.heatmap(corr, annot=True, linewidths=1, cmap="viridis", fmt=".2f")
```
